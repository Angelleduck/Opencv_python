```python
# import of modules
import cv2
import matplotlib.pyplot as plt
```

```python
# loading images
nadia=cv2.imread('img/Nadia_Murad.jpg')
solvay=cv2.imread('img/solvay_conference.jpg')
```

```python
# converting to normal color
nadia=cv2.cvtColor(nadia,cv2.COLOR_BGR2RGB)
solvay=cv2.cvtColor(solvay,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(solvay)
```

```python
# implementing the algorithm 
# Upper c for cascade and Classifer
cascade=cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_default.xml')
```

```python
# implementing the detection

def detect_face(img):
    # making a copy of the image
    face_img = img.copy()
    # detect the position of the face to draw square
    corner=cascade.detectMultiScale(face_img,scaleFactor=1.2)
    
    for(x,y,w,h)in corner:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,200,250),5)
    return(face_img)
```

```python
result=detect_face(solvay)
plt.imshow(result)
```
